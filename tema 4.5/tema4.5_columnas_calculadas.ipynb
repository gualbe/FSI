```python
import pandas as pd
```

### Un data frame con datos de ventas y costes mensuales (meses de enero a junio)

```python
df = pd.DataFrame({
     "Mes":    ["Enero","Febrero","Marzo","Abril","Mayo","Junio"],
     "Ventas": [1000,1200,900,450,1500,300],
     "Costes": [850,1000,1250,500,900,600]
})
```

```python
df
```

### Crear una columna con el margen entre ventas y costes

```python
df['Margen'] = df.Ventas - df.Costes
```

```python
df
```

### Crear una columna con las ventas del mes anterior

```python
df['Ventas_MesAnterior'] = df.Ventas.shift(1)
```

```python
df
```

### Crear dos columnas con las ventas y costes acumulados

```python
df['Ventas_acum'] = df.Ventas.cumsum()
df['Costes_acum'] = df.Costes.cumsum()
```

```python
df
```

### Crear una columna con el margen acumulado (método 1)

```python
df['Margen_acum'] = df.Ventas_acum - df.Costes_acum
```

```python
df
```

### Crear una columna con el margen acumulado (método 2)

```python
df['Margen_acum'] = df.Margen.cumsum()
```
